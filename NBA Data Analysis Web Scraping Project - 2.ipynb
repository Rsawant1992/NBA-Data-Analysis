```python
import pandas as pd
```

```python
mvps = pd.read_csv("mvps.csv")
```

```python
mvps = mvps[["Player", "Year", "Pts Won", "Pts Max", "Share"]]
```

```python
mvps
```

```python
# Cleaning the Player Data
```

```python
players = pd.read_csv("players.csv")
```

```python
players
```

```python
del players["Unnamed: 0"]
```

```python
del players["Rk"]
```

```python
players.head(50)
```

```python
players["Player"].head(50)
```

```python
players["Player"] = players["Player"].str.replace("*", "", regex=False)
```

```python
players.head(45)
```

```python
players.groupby(['Player', 'Year']).get_group(("Greg Anderson", 1991))
```

```python
def single_row(df):
    if df.shape[0]==1:
        return df
    else:
        row = df[df["Tm"] == "TOT"]
        row["Tm"] = df.iloc[-1:]["Tm"]
        return row
    
players = players.groupby(['Player', 'Year']).apply(single_row)
           
```

```python
players.head(20)
```

```python
players.index = players.index.droplevel()
```

```python
players.index = players.index.droplevel()
```

```python
players[players["Player"] == "Greg Anderson"]
```

```python
# Combining the Player and MVP Data
```

```python
combined = players.merge(mvps, how="outer", on=["Player", "Year"])
```

```python
combined.head(5)
```

```python
combined[combined["Pts Won"] > 0]
```

```python
combined[["Pts Won", "Pts Max", "Share"]] = combined[["Pts Won", "Pts Max", "Share"]].fillna(0)
```

```python
combined
```

```python
# Cleaning the Data
```

```python
teams = pd.read_csv("teams.csv")
```

```python
teams
```

```python
teams.head(30)
```

```python
teams = teams[~teams["W"].str.contains("Division")]
```

```python
teams["Team"] = teams["Team"].str.replace("*", "", regex=False)
```

```python
teams.head(15)
```

```python
read_file = pd.read_csv (r'C:\Users\rehan\OneDrive - UMBC\IS 700 IS\Projects\nicknames.txt')
read_file.to_csv (r'C:\Users\rehan\OneDrive - UMBC\IS 700 IS\Projects\nicknames.csv', index=None)
```

```python
nicknames = {}
with open("nicknames.csv") as f:
    lines = f.readlines()
    for line in lines[1:]:
        abbrev,name = line.replace("\n","").split(",")
        nicknames[abbrev] = name
```

```python
combined["Team"] = combined["Tm"].map(nicknames)
```

```python
combined.head()
```

```python
stats = combined.merge(teams, how="outer",on=["Team", "Year"])
```
